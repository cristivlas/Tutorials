```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def gaussian(x, sig, mu=0):
    return 1./(np.sqrt(2.*np.pi)*sig)*np.exp(-np.power((x - mu)/sig, 2.)/2)

N = 100*np.pi
x = np.linspace(-N, N, 1000)
S = 100
y = S * gaussian(x, S) * np.cos(x/np.pi/8)
fig = plt.figure(figsize=(6, 3))
ax = fig.add_subplot()
ax.plot(x, y)
ax.plot(x, S * gaussian(x, S), linestyle='dashed')
ax.plot(x, -S * gaussian(x, S), linestyle='dashed')
ax.grid(True)
plt.show()
```

```python
N = 60*np.pi
x, y = np.meshgrid(np.linspace(-N, N, 500), np.linspace(-N, N, 500))

def radius(x, y):
    return np.sqrt(x**2 + y**2)

z = np.cos(radius(x, y)/np.pi/8) * gaussian(radius(x, y), 100)

fig = plt.figure(figsize=(20, 8))
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(x, y, z, cmap='BuGn')
plt.show()
```

```python
N = 12 * np.pi
x = np.linspace(-N, N, 1000)
y = np.linspace(-N, N, 1000)

x, y = np.meshgrid(x, y)
z = (np.cos(0.5*x/np.pi) + np.cos(0.25*x/np.pi)) * np.sin(0.25*y/np.pi)
w = np.cos(0.25*x/np.pi) * np.cos(0.25*y/np.pi) * np.cos(np.sqrt((0.25*x/np.pi)**2 + (0.25*y/np.pi)**2))**2

fig = plt.figure(figsize=(12, 6))

ax = fig.add_subplot(121, projection='3d'), fig.add_subplot(122, projection='3d')
for a in ax:
    a.set_xlabel('x')
    a.set_ylabel('y')
    a.view_init(elev=30, azim=120)

ax[0].plot_surface(x, y, z, cmap='BuGn')
ax[1].plot_surface(x, y, w, cmap='GnBu')

plt.tight_layout()
plt.show()
```
