```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Evenly spaced numbers over a specified interval:
x = np.linspace(0, 99, num=100, dtype=int) # X axis, 100 integer "steps"
x
```

```python
y = np.random.randint(0, 20, size=100) # Y axis, 100 random integers [0, 20)
y
```

```python
plt.plot(x, y)
plt.axhline(np.std(y), color='green', linestyle='dashed') # draw horiz. line at std dev
plt.grid(True)
plt.show()
```
